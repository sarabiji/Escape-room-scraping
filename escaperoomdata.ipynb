```python
import json

from sc_google_maps_api import ScrapeitCloudClient
client = ScrapeitCloudClient(api_key='023b29ef-7d30-492a-86c7-14727cd0aee9')
response = client.scrape(
  params = {"keyword": "escaperooms","country": "BR","domain": "com"
}
)
#this is the scraped details
#print(response.text)




res = json.loads(response.text)

j =res.get('scrapingResult')
#print(j)
#print(res.get('scrapingResult'))
#for i in j:
#  print(i)


#print(j.get('locals')) ---- its a list



Escape_Room_name =[]
Address =[]
Phone_Number =[]
URL =[]
Hours_of_Operation = []
Reviews =[]
Service_options =[]


k = j.get('locals')
# location of data is in m
for i in k:
  for m in i:
    #print(m)
    s = i.get("title")
    t = i.get("address")
    u = i.get("phone")
    v = i.get("website")
    x = i.get("reviews")
    y = i.get("serviceOptions")
    z = i.get("workingHours")




  #print (s,t,u,v,x,y,z)
  Escape_Room_name.append(s)
  Address.append(t)
  Phone_Number.append(u)
  URL.append(v)
  Hours_of_Operation.append(z)
  Reviews.append(x)
  Service_options.append(y)


#for i in Hours_of_Operation:
#  print(i)
#  #more for loop needed in this area

```

Below code is used to convert to data into a csv file

```python

import csv
with open('escape.csv', 'w+') as myfile:
    writer = csv.writer(myfile)
    writer.writerow(("Escape Room Name", "Address", "Phone","URL","Hours of operation","Reviews","Service_options"))
    for i in range(len(Escape_Room_name)):
        writer.writerow((Escape_Room_name[i], Address[i], Phone_Number[i],URL[i],Hours_of_Operation[i],Reviews[i],Service_options[i]))
```
